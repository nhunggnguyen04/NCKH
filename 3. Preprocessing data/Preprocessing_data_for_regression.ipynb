```python
# Import thư viện cần thiết
import pandas as pd
```

# Nhập dữ liệu và tiền xử lý dữ liệu

```python
# Lấy dataset về bình luận đã gán nhãn và tính điểm cảm xúc của từng sản phẩm
df = pd.read_csv('/content/tiki_reviews_with_sentiment_final.csv')
df.info()
```

```python
label_cols = ['as_content_label_id', 'as_physical_label_id', 'as_price_label_id', 'as_packaging_label_id', 'as_delivery_label_id', 'as_service_label_id','sentiment_llm_label_id']

def get_sentiment_counts(df, col):
    # Pivot table to count occurrences of each label (0, 1, 2)
    counts = df.groupby(['product_id', col]).size().unstack(fill_value=0)
    # Rename columns to reflect the aspect and sentiment
    counts.columns = [f'{col}_{sent}' for sent in counts.columns]
    return counts

# Aggregate all label columns
agg_results = []
for col in label_cols:
    agg_results.append(get_sentiment_counts(df, col))

# Combine all aggregated results
df_aspect_sentiment = pd.concat(agg_results, axis=1).reset_index()

display(df_aspect_sentiment.head())
df_aspect_sentiment.info()
```

```python
# Lấy dataset về dữ liệu sản phẩm
df_prod=pd.read_csv('/content/tiki_product.csv')
display(df_prod)
```

```python
# Hợp nhất dữ liệu vào cùng một bảng
merged_df = df_prod.merge(df_aspect_sentiment, left_on='ID', right_on='product_id', how='left')
display(merged_df.head(5))
```

```python
merged_df.to_csv('merged_df.csv', index=False)
```
